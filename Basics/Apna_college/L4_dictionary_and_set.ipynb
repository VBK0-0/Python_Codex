## Dictionary and set in Python
----------------------------

### Dictionary
------------

- ictionaries are used to store data values in Key: value pairs.
- They are unordered, mutable (changable) & don't allow duplicate keys.

```python
info = {
    "Key": "value",
    "AnotherKey": "AnotherValue",
    "Key3": 3,
    "Key4": [1, 2, 3, 4],
    "Key5": {"NestedKey": "NestedValue"},
    "Key6": True,
    "Key7": (5, 6, 7),
}

print(info)
print(type(info))
print(info["Key"])
print(info["AnotherKey"])
print(info["Key3"])
print(info["Key4"])
print(info["Key5"])
print(info["Key6"])
print(info["Key7"])
print(info["Key5"]["NestedKey"])
print(type(info["Key7"]))

# Updating values and adding new key-value pairs
info["Key3"] = 42
info["NewKey"] = "NewValue"
print(info)
```

```python
# Null dictionary
null_dict = {}
print(null_dict)
```

```python
# Dictionary methods
info = {
    "Key": "value",
    "AnotherKey": "AnotherValue",
    "Key3": 3,
    "Key4": [1, 2, 3, 4],
    "Key5": {"NestedKey": "NestedValue"},
    "Key6": True,
    "Key7": (5, 6, 7),
}

print(info.keys())      #returns all keys as list-like object
print(info.values())    #returns all values
print(info.items())   #returns all key-value pairs as tuples
print(info.get("Key3"))  #returns value for the given key
print(info.get("NonExistentKey", "DefaultValue"))  #returns default value if key is not found
print(info.pop("Key4"))  #removes and returns the value for the given key
print(info.update({"Key8": "Value8", "Key9": "Value9"}))  #updates the dictionary with new key-value pairs
print(info)
print(len(info))  #returns the number of key-value pairs in the dictionary  
```

## Set in Python
-----------------------

- Set is a collection of unordered items.
- Each element in the set must be unique and immutable.

nums = {1, 2, 3, 4}

set2 = {1, 2, 2, 2}

/#repeated elements stored only once, so it resolved to {1, 2}

null_set = set()     #empty set sequence

- Boolean, int, float, str, touple -> Can be used in set (immutable).
- List, dictionary -> Can not be used in set (mutable)


```python
collection = {1, 2, 3, 4}
print(collection)
print(type(collection))
```

```python
set_repeated = {1, 2, 2, 3, 4, 4, 5}
print(set_repeated)  # Output: {1, 2, 3, 4, 5}
```

```python
collections_info = {1, 2, 1, "Hello", "Hello", 3.14, 3.14, (1, 2), (1, 2)}
print(collections_info)  # Output: {1, 2, 'Hello', 3.14, (1, 2)}    
print(len(collections_info))  # Total number o f unique elements in the set
```

```python
empty_dict = {} # This is not a set, it is a dictionary 
print(type(empty_dict)) # Output: <class 'dict'>
empty_set = set() # This is a set, it is created using the set() constructor
print(type(empty_set)) # Output: <class 'set'>
```

```python
# et methods
my_set = set()
my_set.add(1)  # Adds an element to the set
my_set.add(2)
my_set.add(2)  # Duplicate element, will not be added
print(my_set)  # Output: {1, 2}print(my_set) 
my_set.remove(1)  # Removes an element from the set
print(my_set)  # Output: {2}
my_set.discard(3)  # Removes an element if it exists, does nothing if it doesn't
print(my_set)  # Output: {2}
my_set.add(3)
my_set.add(4)
print(my_set)  # Output: {2, 3, 4}
my_set.discard(3)  # Removes an element if it exists, does nothing if it doesn't
print(my_set)  # Output: {2, 4}
my_set.pop()  # Removes and returns an arbitrary element from the set
print(my_set)  # Output: {4} (or) {2}, depending on which element was popped)
my_set.clear()  # Removes all elements from the set
print(my_set)
```

```python
my_set = set()
my_set.add(1)
my_set.add(2)
my_set.add(3)
my_set.add((4, 5))
my_set.add([4, 5])  # This will raise a TypeError because lists are unhashable
print(my_set) 
```

```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
print(set1)
print(set2)
union_set = set1.union(set2)  # or set1 | set2
print("union is:", union_set)  # Output: {1, 2, 3, 4, 5, 6}
print("intersection is:", set1.intersection(set2))  # Output: {3, 4}
print("difference is:", set1.difference(set2))  # Output: {1, 2}
print("symmetric difference is:", set1.symmetric_difference(set2))  # Output: {1, 2, 5, 6}
```

4. Figure out a way to store 9 & 9.0 as seperate values in a set. (You can take help of built-in data types)

```python
values = {9, 9.0}
print(values)  # Output: {9}, because 9 and 9.0 are considered the same in a set

# Possible solutions:
values = {9, "9.0"}
print(values)  # Output: {9, '9.0'}, now 9 and "9.0" are treated as distinct elements in the set

values = {"9", 9.0}
print(values)  # Output: {'9', 9.0}, now "9" and 9.0 are treated as distinct elements in the set

values = {("float", 9.0), ("int", 9)}
print(values)  # Output: {('float', 9.0), ('int', 9)}, now the tuples are treated as distinct elements in the set
```
