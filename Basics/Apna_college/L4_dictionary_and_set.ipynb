## Dictionary and set in Python
----------------------------

### Dictionary
------------

- ictionaries are used to store data values in Key: value pairs.
- They are unordered, mutable (changable) & don't allow duplicate keys.

```python
info = {
    "Key": "value",
    "AnotherKey": "AnotherValue",
    "Key3": 3,
    "Key4": [1, 2, 3, 4],
    "Key5": {"NestedKey": "NestedValue"},
    "Key6": True,
    "Key7": (5, 6, 7),
}

print(info)
print(type(info))
print(info["Key"])
print(info["AnotherKey"])
print(info["Key3"])
print(info["Key4"])
print(info["Key5"])
print(info["Key6"])
print(info["Key7"])
print(info["Key5"]["NestedKey"])
print(type(info["Key7"]))

# Updating values and adding new key-value pairs
info["Key3"] = 42
info["NewKey"] = "NewValue"
print(info)
```

```python
# Null dictionary
null_dict = {}
print(null_dict)
```

```python
# Dictionary methods
info = {
    "Key": "value",
    "AnotherKey": "AnotherValue",
    "Key3": 3,
    "Key4": [1, 2, 3, 4],
    "Key5": {"NestedKey": "NestedValue"},
    "Key6": True,
    "Key7": (5, 6, 7),
}

print(info.keys())      #returns all keys as list-like object
print(info.values())    #returns all values
print(info.items())   #returns all key-value pairs as tuples
print(info.get("Key3"))  #returns value for the given key
print(info.get("NonExistentKey", "DefaultValue"))  #returns default value if key is not found
print(info.pop("Key4"))  #removes and returns the value for the given key
print(info.update({"Key8": "Value8", "Key9": "Value9"}))  #updates the dictionary with new key-value pairs
print(info)
print(len(info))  #returns the number of key-value pairs in the dictionary  
```
