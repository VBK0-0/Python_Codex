### Basic operations
- Concatination

"hello" + "world"

- Length of string

len(str)

```python
str1 = "This is a string"
print(str1)
print(type(str1))
```

NOTE: Escape sequence characters are used for formatting

tab: \t

next line: \n

```python
str1 = 'This is a string.\tThere is a tab space between the two sentences.'
str2 = "This is also a string. \nThis will be printed in a new line"
print(str1)
print(str2)
```

```python
#Concatenation
str1 = "Hello"
str2 = "World"
str3 = str1 + " " + str2
print(str3)
```

```python
#len() function
str1 = "Balakrishna"
str2 = "Varjula Balakrishna"
str3 = " " 
print(len(str1))
print(len(str2))
print(len(str3))
```

## Slicing
- Accessing parts of a string.
- str[starting_idx : ending_idx]  #ending idx is not included 

```python
str1 = "VBK Productions" 
print(str1[0])  # V
print(str1[4])  # P
print(str1[0:3])  # VBK
print(str1[4:])   # Productions
print(str1[:3])   # VBK
print(str1[::2])  # VKPoutos
```

```python
# Negative indexing
str1 = "Apple"
print(str1[-1])  # e
print(str1[-2])  # l
print(str1[-3])  # p
print(str1[-4])  # l
print(str1[-5])  # A    
print(str1[-5:]) # Apple
print(str1[:-3])  # Ap
```

```python
# String Functions
str1 = "I am learning Python"
print(str1)  # I am learning Python
print(str1.endswith("Python"))  # True
print(str1.endswith("python"))  # False
print(str1.startswith("I am"))  # True
print(str1.startswith("i am"))  # False
print(str1.islower())  # False
print(str1.isupper())  # False
print(str1.lower())  # i am learning python
print(str1.upper())  # I AM LEARNING PYTHON
print(str1.count("a"))  # 2
print(str1.count("am"))  # 1
print(str1.count("Python"))  # 1    
print(str1.find("Python"))  # 14
print(str1.find("python"))  # -1
                            # find() function returns the index of the first occurrence of the substring. 
                            # If the substring is not found, it returns -1.
print(str1.replace("Python", "ML"))  # I am learning ML
print(str1)  # I am learning Python
             # Original string is not changed because strings are immutable in Python
str2 = str1.replace("Python", "ML")
print(str2)  # I am learning ML
```

## Conditonal Statements
------------------------------
- if- elif-else(SYNTAX)

if(condition):

    statement1

elif(condition):

    statement2
else:

    statementN

```python
light = "red"
if(light == "red"):
    print("Stop")
elif(light == "yellow"):
    print("Get Ready")
elif(light == "green"):
    print("Go")
else:
    print("Invalid light color")
```

```python
num = 5
if(num > 2):
    print("Greater than 2")
if(num > 3):
    print("Greater than 3")
```

```python
num = 5
if(num > 2):
    print("Greater than 2")
elif(num > 3):
    print("Greater than 3")
```

```python
# Nesting
age = 25
if(age > 18):
    print("You are an adult")
    if(age > 60):
        print("You are a senior citizen")
    else:
        print("You are not a senior citizen")
```

2. WAP to find the greatest of 3 numbers entered by the user.

```python
in1 = int(input("Enter first number: "))
in2 = int(input("Enter second number: "))
in3 = int(input("Enter third number: "))
if(in1 >= in2 and in1 >= in3):
    print("Largest number is: " + str(in1))
elif(in2 >= in1 and in2 >= in3):
    print("Largest number is: " + str(in2))
else:
    print("Largest number is: " + str(in3))
```
