## Strings
- String is a data type that storess a sequence of characters

### Basic operations
- Concatination

"hello" + "world"

- Length of string

len(str)

```python
str1 = "This is a string"
print(str1)
print(type(str1))
```

NOTE: Escape sequence characters are used for formatting

tab: \t

next line: \n

```python
str1 = 'This is a string.\tThere is a tab space between the two sentences.'
str2 = "This is also a string. \nThis will be printed in a new line"
print(str1)
print(str2)
```

```python
#Concatenation
str1 = "Hello"
str2 = "World"
str3 = str1 + " " + str2
print(str3)
```

```python
#len() function
str1 = "Balakrishna"
str2 = "Varjula Balakrishna"
str3 = " " 
print(len(str1))
print(len(str2))
print(len(str3))
```
