## Types of operators:

An operator is a symbol that performs a certain operation between operands.

1. Arithmetic Operators (+, -, *, /, %, **, //)
2. Relational/ Comparision Operators (==, !=, >, <, >=, <=)
3. Assignment Operators ( =, +=, -=, *=, /=, %=, **=, //=)
4. Logical Operators (not, and, or)
5. Bitwise Operators (&, `, ^, ~, <<, >>)
6. Membership Operators (in, not in)
7. Identity Operators (is, not is)

```python
#1. Arithmetic Operators (+, -, *, /, %, **, //)
a = 5
b = 10
print("The sum of a and b is: ", a + b)
print("The difference of a and b is: ", a - b)
print("The product of a and b is: ", a * b)
print("The division of a and b is: ", a / b)
print("The modulus (remainder) of a and b is: ", a % b)
print("The exponent of a and b is: ", a ** b)
print("The floor division of a and b is: ", a // b)
```

```python
#2. Relational/ Comparision Operators (==, !=, >, <, >=, <=)
a = 5
b = 10
print("Is a equal to b? ", a == b)
print("Is a not equal to b? ", a != b)
print("Is a greater than b? ", a > b)
print("Is a less than b? ", a < b)
print("Is a greater than or equal to b? ", a >= b)
print("Is a less than or equal to b? ", a <= b)
```

```python
#3. Assignment Operators ( =, +=, -=, *=, /=, %=, **=, //=)
a = 5
a += 10 # a = a + 10
print("The value of a after addition is: ", a)
a -= 10 # a = a - 10
print("The value of a after subtraction is: ", a)
a *= 10 # a = a * 10
print("The value of a after multiplication is: ", a)
a /= 10 # a = a / 10
print("The value of a after division is: ", a)
a %= 10 # a = a % 10
print("The value of a after modulus is: ", a)
a **= 10 # a = a ** 10
print("The value of a after exponent is: ", a)
a //= 10 # a = a // 10
print("The value of a after floor division is: ", a)
```

```python
#4. Logical Operators (not, and, or)
a = True
b = False
print("The value of a is: ", a)
print("The value of b is: ", b)
print("The value of not a is: ", not a)
print("The value of a and b is: ", a and b)
print("The value of a or b is: ", a or b)
```

```python
#5. Identity Operators (is, is not)
a = 5
b = 5
print("The value of a is: ", a)
print("The value of b is: ", b)
print("Is a the same object as b? ", a is b)
b = 10
print("The value of b is: ", b)
print("Is a the not same object as b? ", a is not b)
```

```python
#6. Membership Operators (in, not in)
a = [1, 2, 3, 4, 5]
print("The value of a is: ", a)
print("Is 3 in a? ", 3 in a)
print("Is 6 not in a? ", 6 not in a)    
```

```python
#7. Bitwise Operators (&, |, ^, ~, <<, >>)
a = 5 # 0b0101
b = 3 # 0b0011
print("The value of a is: ", a)
print("The value of b is: ", b)
print("a & b = ", a & b)
print("a | b = ", a | b)
print("a ^ b = ", a ^ b)
print("~a = ", ~a)
print("a << 1 = ", a << 1)
print("a >> 1 = ", a >> 1)
```

In python:
1. Type conversion - interpreter automatically converts
2. Type casting - done manually by programmer

```python
#Type Conversion
a = 2 # Integer
b = 4.5 # Floating point number
sum = a + b # The result will be a floating point number because of the presence 
            #of a floating point number in the expression.
            # Floating point numbers have higher precedence than integers in arithmetic operations.
print("The sum of a and b is: ", sum)
print(type(sum))
```

```python
a, b = 1, "2"
sum = a + b # This will raise a TypeError because we cannot add an integer and a string.    
print("Sum: ", sum)
```

```python
#Type Casting
a, b = 1, "2"
c = int(b) # Type casting the string "2" to an integer 2
sum = a + c     
print(type(c))   
print("Sum: ", sum)
```

```python
a = "Balakrishna"
b = int(a) # This will raise a ValueError because the string "Balakrishna" cannot be converted to an integer.   
print("Sum: ", b)
```

```python
a = 3.14
b =str(a)
print(b)
print(type(b))
```
