# Lists and Tuples in Python
---------------------------------------
## Lists in Python:
- A built-in data type that store set of values.
- It can store elements of different types (integer, float, string, etc.)

marks = [87, 64, 33, 95, 76] #marks[0], marks[1]..

student = ["Kiran", 85, "Delhi"]  #student[0], student[1]..

student[0] = "Arjun"  #Allowed in Python

len(student)   #returns length

```python
marks = [85, 90, 78, 92, 88]
print("Marks:", marks)
print(type(marks))
print("Number of marks:", len(marks))
print("First mark:", marks[0])
print("Last mark:", marks[-1])
```

NOTE: In Python, the concepts of strings and lists are similar with the difference being that strings are immutable, where as lists are mutable.

Strings -> Immutable

Lists -> Mutable

```python
# Lists are mutable, we can change their elements
student =["VBK0-0", 24, "ECE"]
print("Student:", student)
student[0] = "Varjula Balakrishna"
print("Updated Student:", student)
```

## List Slicing
- Similar to string slicing

list_name[starting_idx : ending_idx] #ending idx is not included

```python
marks = [87, 64, 33, 95, 76]
print("Marks:", marks)
print(marks[1:4])  # Slicing the list
print(marks[:4])   # First four marks, same as marks[0:4]
print(marks[2:])   # Marks from index 2 to the end, same as marks[2:len(marks)]
print(marks[-3:-1])  # Marks from index -3 to -1
```

```python
# List methods
fruits = ["apple", "banana", "cherry"]
print("Fruits:", fruits)
fruits.append("orange")  # Add an element to the end of the list
print("After append:", fruits)
fruits.insert(1, "grape")  # Insert an element at index 1
print("After insert:", fruits)
fruits.remove("banana")  # Remove the first occurrence of "banana"
print("After remove:", fruits)
fruits.pop()  # Remove and return the last element
print("After pop:", fruits)
fruits.sort()  # Sort the list in place
print("After sort:", fruits)
fruits.reverse()  # Reverse the list in place
print("After reverse:", fruits)
fruits.clear()  # Remove all elements from the list
print("After clear:", fruits)
fruits.extend(["kiwi", "melon", "peach"])  # Extend the list with another list
print("After extend:", fruits)
fruits.index("melon")  # Get the index of "melon"
print("Index of melon:", fruits.index("melon"))
fruits.count("kiwi")  # Count the occurrences of "kiwi"
print("Count of kiwi:", fruits.count("kiwi"))
fruits.copy()  # Create a shallow copy of the list
print("Copy of fruits:", fruits.copy())
```

## Tuples in Python
----------------------------

- A built-in datatype that  let's us create immutable sequences of values. 
- Tuples are immutable.

```python
tup1 = (87, 64, 33, 95, 76) # tup[0], tup[1], tup[2], tup[3], tup[4]
print("Tuple:", tup1)
print(type(tup1))
print("Number of elements in tuple:", len(tup1))

tup2 = ()
print("Empty Tuple:", tup2)
 
tup2 = (42,) # A single element tuple must have a comma
print("Single element Tuple:", tup2)

tup[0] = 43 # This will raise a TypeError because tuples are immutable
```

```python
# Tuple methods
tup1 = (2, 1, 3, 1)
print("Tuple:", tup1)
print("Count of 1:", tup1.count(1))  # Count the occurrences of 1
print("Index of 1:", tup1.index(1))  # Get the index of the first occurrence of 1
```

2. WAP to check if a list contains a palindrome of elements. (Hint: Use copy method)

```python
list1 = [1, 2, 3, 2, 1]
list2 = list1.copy()
list2.reverse()  # Create a copy of list1
print("Original list:", list1)
print("Reversed copy:", list2)
if list1 == list2:
    print("The given list is a palindrome.")
else:
    print("The given list is not a palindrome.")
```
