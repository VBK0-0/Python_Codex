# Lists and Tuples in Python
---------------------------------------
## Lists in Python:
- A built-in data type that store set of values.
- It can store elements of different types (integer, float, string, etc.)

marks = [87, 64, 33, 95, 76] #marks[0], marks[1]..

student = ["Kiran", 85, "Delhi"]  #student[0], student[1]..

student[0] = "Arjun"  #Allowed in Python

len(student)   #returns length

```python
marks = [85, 90, 78, 92, 88]
print("Marks:", marks)
print(type(marks))
print("Number of marks:", len(marks))
print("First mark:", marks[0])
print("Last mark:", marks[-1])
```

NOTE: In Python, the concepts of strings and lists are similar with the difference being that strings are immutable, where as lists are mutable.

Strings -> Immutable

Lists -> Mutable

```python
# Lists are mutable, we can change their elements
student =["VBK0-0", 24, "ECE"]
print("Student:", student)
student[0] = "Varjula Balakrishna"
print("Updated Student:", student)
```
