## Strings

```python
phrase = "Giraffe Academy"
#         0123456789111315
#                  101214
print(phrase + " is cool")
print(phrase.lower())
print(phrase.upper())
print(phrase.isupper())
print(phrase.islower())
print(phrase.upper().isupper())
print(len(phrase))
print(phrase[0])
print(phrase[3])
print(phrase[0:5])
print(phrase.index("G"))
print(phrase.replace("Giraffe", "Elephant"))
```
